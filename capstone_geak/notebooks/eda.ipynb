```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("data/retail_price.csv")

# View first 5 rows
data.head()
```

```python
# Shape of dataset
data.shape
```

```python
# Info about dataset
data.info()
```

Data Dictionary

Column	Description
product_id	Unique product identifier
product_category_name	Category of product
month_year	Month and year of sales
qty	Quantity sold (Demand)
total_price	Total revenue (qty × unit price)
freight_price	Shipping cost
unit_price	Price per unit
product_name_lenght	Length of product name
product_description_lenght	Description length
product_photos_qty	Number of product images

```python
# Check missing values
data.isnull().sum()
```

```python
data.describe()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='unit_price', y='qty', data=data)
plt.title("Unit Price vs Quantity Sold")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='unit_price', y='total_price', data=data)
plt.title("Price vs Revenue")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='product_category_name', data=data)
plt.xticks(rotation=90)
plt.title("Product Category Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='freight_price', y='qty', data=data)
plt.title("Freight Price vs Demand")
plt.show()
```

```python
data['month_year'] = pd.to_datetime(data['month_year'])

monthly_sales = data.groupby('month_year')['qty'].sum()

monthly_sales.plot(figsize=(10,5))
plt.title("Monthly Sales Trend")
plt.show()
```
